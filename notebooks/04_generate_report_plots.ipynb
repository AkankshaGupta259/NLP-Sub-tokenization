```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Ensure the figures directory exists
os.makedirs('../reports/figures/', exist_ok=True)

# Set global styling for professional academic plots
plt.style.use('ggplot')
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

print("Generating plots for the final report...")

# ==========================================
# 1. Performance Metrics Plot
# ==========================================
perf_path = '../reports/performance_metrics.csv'
if os.path.exists(perf_path):
    df_perf = pd.read_csv(perf_path)
    
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle('BPE Performance Scaling', fontsize=16, fontweight='bold', y=1.05)
    
    # Train Time
    axes[0].plot(df_perf['Vocab Size'], df_perf['Time (s)'], marker='o', color=colors[0], linewidth=2.5, markersize=8)
    axes[0].set_title('Training Time vs. Vocab Size')
    axes[0].set_xlabel('Vocabulary Size')
    axes[0].set_ylabel('Time (seconds)')
    
    # Compression Ratio
    axes[1].plot(df_perf['Vocab Size'], df_perf['Compression Ratio'], marker='s', color=colors[1], linewidth=2.5, markersize=8)
    axes[1].set_title('Compression Ratio vs. Vocab Size')
    axes[1].set_xlabel('Vocabulary Size')
    axes[1].set_ylabel('Compression Ratio')
    
    # Tokens per Word
    axes[2].plot(df_perf['Vocab Size'], df_perf['Avg Tokens/Word'], marker='^', color=colors[2], linewidth=2.5, markersize=8)
    axes[2].set_title('Fragmentation vs. Vocab Size')
    axes[2].set_xlabel('Vocabulary Size')
    axes[2].set_ylabel('Avg Tokens / Word')
    
    plt.tight_layout()
    plt.savefig('../reports/figures/01_performance_scaling.png', dpi=300, bbox_inches='tight')
    plt.close()
    print(" Saved Performance plot: reports/figures/01_performance_scaling.png")
else:
    print(f" Could not find {perf_path}")

# ==========================================
# 2. Robustness Under Noise Plot
# ==========================================
rob_path = '../reports/robustness_metrics.csv'
if os.path.exists(rob_path):
    df_rob = pd.read_csv(rob_path)
    
    fig, ax1 = plt.subplots(figsize=(10, 6))
    
    # Bar chart for Token Count
    bars = ax1.bar(df_rob['Condition'], df_rob['Token Count'], color=['#3498db', '#e74c3c', '#9b59b6'], alpha=0.8)
    ax1.set_title('Robustness: Tokenization Length Under Noisy Conditions', fontsize=14, fontweight='bold')
    ax1.set_ylabel('Total Token Count', fontsize=12, fontweight='bold')
    ax1.set_xlabel('Input Condition', fontsize=12)
    
    # Add value labels on top of bars
    for bar in bars:
        yval = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2, yval + 1, int(yval), ha='center', va='bottom', fontweight='bold')

    plt.tight_layout()
    plt.savefig('../reports/figures/02_robustness_noise.png', dpi=300, bbox_inches='tight')
    plt.close()
    print(" Saved Robustness plot: reports/figures/02_robustness_noise.png")
else:
    print(f" Could not find {rob_path}")

# ==========================================
# 3. Linguistic Fairness Audit Plot
# ==========================================
fair_path = '../reports/fairness_audit.csv'
if os.path.exists(fair_path):
    df_fair = pd.read_csv(fair_path)
    
    fig, ax1 = plt.subplots(figsize=(10, 6))
    
    # Plotting Fragmentation (Tokens per Word)
    bars = ax1.bar(df_fair['Language / Group'], df_fair['Tokens per Word'], color=['#2ecc71', '#e67e22', '#f1c40f'], edgecolor='black')
    ax1.set_title('Linguistic Fairness: Token Fragmentation by Language', fontsize=14, fontweight='bold')
    ax1.set_ylabel('Tokens per Word (Fragmentation)', fontsize=12, fontweight='bold')
    
    # Highlight the disparity visually
    ax1.axhline(y=df_fair['Tokens per Word'].iloc[0], color='r', linestyle='--', alpha=0.5, label='English Baseline')
    ax1.legend()
    
    # Add value labels
    for bar in bars:
        yval = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2, yval + 0.1, round(yval, 2), ha='center', va='bottom', fontweight='bold')

    plt.tight_layout()
    plt.savefig('../reports/figures/03_fairness_audit.png', dpi=300, bbox_inches='tight')
    plt.close()
    print(" Saved Fairness plot: reports/figures/03_fairness_audit.png")
else:
    print(f" Could not find {fair_path}")

print("\n All plots successfully generated! Check the reports/figures/ directory.")
```
